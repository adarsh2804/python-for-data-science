### Problem-1:
**WAP that will accept two sets of numbers and calculate the Pearson’s correlation coefficient between them. (Use necessary exception handling). (Optional if you do not know what it is)**
<img src="https://www.statisticshowto.com/wp-content/uploads/2012/10/pearson.gif" width="300" height="100">

```python
# Solution-1

try:
    sum_of_x = 0
    sum_of_y = 0
    sum_of_xy = 0
    square_sum_of_x = 0
    square_sum_of_y = 0

    x = [10, 30, 35, 68, 70]                                # input is given in the program itself
    y = [25, 28, 37, 43, 58]            

    n = len(x)                                              # gives length of list x

    for i in range(0, n) :
        
        sum_of_x = sum_of_x + x[i]                          # sum of elements of list x
        sum_of_y = sum_of_y + y[i]                          # sum of elements of list y
        sum_of_xy = sum_of_xy + (x[i] * y[i])               # sum of x[i] * y[i]

        square_sum_of_x = square_sum_of_x + (x[i] * x[i])   # sum of square of elements of list x
        square_sum_of_y = square_sum_of_y + (y[i] * y[i])   # sum of square of elements of list y

        # using formula for calculating correlation coefficient
        correlation = ((n * sum_of_xy) - (sum_of_x * sum_of_y)) / ((((n * square_sum_of_x) - sum_of_x**2) * ((n * square_sum_of_y) - sum_of_y**2))**(1/2))

    print(correlation)
    
except:
    print("Some error occured")                             
```

### Problem-2:
**Write a program to print all prime numbers from 1 to 300.**

```python
# Solution-2

for i in range(2, 301):               # since 1 is not prime number, range starts from 2 to 300
    count = 0
    for j in range(2, (i//2)+1):      # if number is not divided by any number in a range from 2 to its half (i//2), it is a prime number
        if i % j == 0:                # if number is divided by any number in range from 2 to its half, its not prime
            count += 1
            break                     # it wont check further since the number is not prime
    if count == 0:                    
        print(i)                      # i is a prime number
```

### Problem-4:
**WAP which repeatedly reads numbers until the user enters “done”. Once “done” is entered, print out the total, count, and average of the numbers. If the user enters anything other than a number, detect their mistake using try and except and print an error message and skip to the next number. (Use of list is not allowed in this program)**
 
Enter a number: 4

Enter a number: 5

Enter a number: nine

Invalid input

Enter a number: 7

Enter a number: done

<pre>16	3	5.33333333333</pre>

```python
# Solution-4

count = 0
total = 0

while True:
    try:
        num = input("Enter a number : ")
        if num == "done":
            break                                    # it will terminate the loop once user enters done
        else:
            total += int(num)                        # number is added to total in each iteration
            count += 1                               # for each iteration count is increased
    except:
        print("Input should be numeric")             # exception will be raised if input is non-numeric

average = total / count
print("total =", total)
print("count =", count)
print("average =", average)
```

### Problem-5:
**Write another program that prompts for a list of numbers as above and at the end prints both the maximum and minimum of the numbers instead of the average. (Use of list is not allowed in this program)**

```python
# Solution-5

count = 0
total = 0

while True:
    try:
        num = input("Enter a number : ")
        if num == "done":
            break                           # it will terminate the loop once user enters done
        else:
            total += int(num)               # number is added to total in each iteration
            count += 1                      # for each iteration count is increased
            
            if count == 1:
                minimum = int(num)          # it will set the first value as minimum value
                maximum = int(num)          # it will set the first value as maximum value
            
            if int(num) < minimum:          # comparing the value of each iteration with the minimum value
                minimum = int(num)          # if the value in this iteration is less than minimum value, it will be set as minimum
            
            if int(num) > maximum:
                maximum = int(num)          # if the value in this iteration is more than maximum value, it will be set as maximum
            
    except:
        print("Input should be numeric")    # exception will be raised if input is non-numeric

print("total =", total)
print("count =", count)
print("Minimum value =", minimum)
print("Maximum value =", maximum)
```

### Problem-6:
**Guess a number is a game that prompts a player to guess a number between 0 and 9, which is randomly generated by the system. When the input given by the user matches the number generated by the system then the user wins. The game should go as follows:**
	
	Guess the number: 5
	Sorry, try again
	Guess the number: 3
	Sorry, try again
	Guess the number: 8
	You got it right! Congo!

```python
# Solution-6

import random

number = random.randint(0, 9)                             # it will generate a random integer between 0 to 9

while True:
    try:
        guess = int(input("Guess the number : "))         # taking input as int type
        
        if guess < 0 or guess > 9:
            print("Pls enter the integer between 0 to 9")
        elif guess == number :
            print("You got it right! Congo!")             # when the guess is right loop will be terminated
            break
        else:
            print("Sorry, try again")                     # when guess is not right another chance will be given
            
    except:
        print("Enter a numeric value")                    # Exception will be raised for non-numeric input
```

### Problem-7:
**Make an improvement to the Guess a number game. Guide the user where they are standing and limit the number of attempts to 3. For example, the game should go like this:**
	
    Guess the number: 4
    Too low
    Guess the number: 9
    Too high
    Guess the number: 5
    Sorry you Lost!

```python
# Solution-7

import random

counter = 0
number = random.randint(0, 9)                             # it will generate a random integer between 0 to 9

while counter < 3:                                        # user will be given maximum 3 chances
    try:
        guess = int(input("Guess the number : "))         # taking input as int type
        
        if guess < 0 or guess > 9:
            print("Pls enter the integer between 0 to 9")
            continue                                      # since input is out of range it will skip rest of the steps in loop
            
        if guess == number :
            print("You got it right! Congo!")             # when the guess is right loop will be terminated
            break
        
        counter += 1                                      # keeps track of number of chances given
        difference = guess - number                       # it will be used to give hint to the user
        
        if counter < 3:                                   # The hints should be given for only 2 wrong attempts
            if difference >= 5:
                print("Too high")
            elif difference > 0:
                print("High")
            elif difference > -5:
                print("Low")
            else:
                print("Too low")
        else:
            print("Sorry, you lost!")                    # if the third attempt is wrong, loop will be terminated
            break
            
    except:
        print("Enter a numeric value")                   # Exception will be raised for non-numeric input
```

### Problem-8:
**Let us make the above game a little more interesting by converting it into a gambling problem. Suppose that a player starts with Rs. 1,000. If he loses then he will lose Rs. 500. For example the game should go like this:**

    You have a cash of Rs. 1,000 with you... 
    Guess the number: 8

```python
# Solution-8

import random

counter = 0
balance = 1000
number = random.randint(0, 9)                             # it will generate a random integer between 0 to 9

while counter < 3:                                        # user will be given maximum 3 chances
    try:
        if counter == 0:
            print("You have a cash of Rs. 1,000 with you...")
        
        guess = int(input("Guess the number : "))         # taking input as int type
        
        if guess < 0 or guess > 9:
            print("Pls enter the integer between 0 to 9")
            continue                                      # since input is out of range it will skip rest of the steps in loop
            
        if guess == number :
            if counter == 0:
                balance += 5000
                print("You have just won Rs. 5,000")
                print("Your balance : Rs.", balance)
            elif counter == 1:
                balance += 1000
                print("You have just won Rs. 1,000")
                print("Your balance : Rs.", balance)
            else:
                print("You got it right! Congo!")         # when the guess is right loop will be terminated
                print("Your balance : Rs.", balance)
            break
        
        counter += 1                                      # keeps track of number of chances given
           
        if counter == 3:                                  # The hints should be given for only 2 wrong attempts
            print("You have lost Rs. 500")                # if the third attempt is wrong, loop will be terminated
            balance -= 500
            print("Your balance : Rs.", balance)
            break
        
        difference = guess - number                       # it will be used to give hint to the user
     
        if difference >= 5:
            print("Too high")
        elif difference > 0:
            print("High")
        elif difference > -5:
            print("Low")
        else:
            print("Too low")
            
    except:
        print("Enter a numeric value")                    # Exception will be raised for non-numeric input
```

```python
# Solution-9

import random

counter = 0
flag = 0                                                  # This variable will be set 1 after 1 set of game
balance = 500
ewallet_balance = 500
answer = ''

while counter <= 3:                                       # user will be given maximum 3 chances in one set of game
    try:
        if counter != 3:
            if counter == 0:
                number = random.randint(0, 9)             # it will generate a random integer between 0 to 9 when new game starts
                flag = 0                        
                print("You have a cash of Rs. %d with you..." %balance)   # formatting is done to show the balance before starting the game


            guess = int(input("Guess the number : "))     # taking input from user

            if guess < 0 or guess > 9:
                print("Pls enter the integer between 0 to 9")
                continue                                  # since input is out of range it will skip rest of the steps in loop

            if guess == number :                          # when user guess the correct number
                if counter == 0:
                    balance += 5000
                    print("You have just won Rs. 5,000")  # if user guesses the correct number in first attempt, reward is 5000
                    print("Your balance : Rs.", balance)
                    flag = 1                              # flag is set to 1 to give user another chance to play the game
                elif counter == 1:
                    balance += 1000
                    print("You have just won Rs. 1,000")  # if user guesses correct number in second attempt, reward is 1000
                    print("Your balance : Rs.", balance)
                    flag = 1
                else:
                    print("You got it right! Congo!")     # if user guesses correct number in third attempt, no reward
                    print("Your balance : Rs.", balance)
                    flag = 1

            else:
                difference = guess - number               # it will be used to give hint to the user

                if difference >= 5:
                    print("Too high")
                elif difference > 0:
                    print("High")
                elif difference > -5:
                    print("Low")
                else:
                    print("Too low")

                counter += 1                              # keeps track of number of chances given
                
        if counter == 3 and flag == 0:                    # The hints should be given for only 2 wrong attempts
            print("You have lost Rs. 500")                # if the third attempt is wrong, loop will be terminated
            balance -= 500
            print("Your balance : Rs.", balance)
            flag = 1
            
        if flag == 1:                                     
            answer = input("Would you like to play one more time? (Yes/No) : ")
        
            if answer == 'No':
                print("Have a wonderful day!")            # User doesn't want to play more so break the loop
                break
        
            elif answer == 'Yes':
                
                while balance < 500:                      # For balance less than 500, it will prompt user to withdraw money from e-wallet
                    print("You have to withdraw money from your e-wallet")
                    amount = int(input("Enter the amount to withdraw from e-wallet : "))
                    ewallet_balance -= amount
                    
                    if ewallet_balance < 0:
                        print("No sufficient amount in e-wallet. You cannot play")
                        break
                    
                    balance += amount
                    counter = 0
                    
                    if balance >= 500:                    # it will restart the game for user
                        counter = 0
                        break

                if ewallet_balance < 0:
                    break
                else:
                    counter = 0
                    continue

            else:
                print("Input should be either Yes or No!")
                flag = 1

    except:
        print("Enter a numeric value")                    # Exception will be raised for non-numeric input
```
